```python
import pandas as pd
import numpy as np

players_correct_ids = pd.read_csv('stats/player_table.csv')
players_final = pd.read_csv('stats/Player_Final.csv')
end_of_seasons = pd.read_csv('stats/end_of_season_teams.csv')
```

```python
# fix player ids in players_final
player_ids = {}
for i, player in enumerate(players_correct_ids['player']):
    player_ids[player] = i + 1

players_final['new_player_id'] = players_final['player'].map(player_ids).fillna(0).astype('Int64')
players_final
```

```python
# fix season id in players final  
for i in range(len(end_of_seasons)):
    seas_id = players_final.loc[players_final['seas_id'] == end_of_seasons['seas_id'][i]]
    player_id = players_final.loc[players_final['seas_id'] == end_of_seasons['seas_id'][i], 'new_player_id']
    
    end_of_seasons.loc[i, 'seas_id'] = seas_id['season'].values[0]
    end_of_seasons.loc[i, 'player_id'] = player_id.values[0]

end_of_seasons
```

```python
# create new end of seasons table with correct ids
years = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
end_of_seasons = end_of_seasons.loc[end_of_seasons['seas_id'].isin(years)]

for i, year in enumerate(end_of_seasons['seas_id']):
    if year == 2023:
        end_of_seasons.loc[i, 'seas_id'] = 7
    if year == 2022:
        end_of_seasons.loc[i, 'seas_id'] = 6
    if year == 2021:
        end_of_seasons.loc[i, 'seas_id'] = 5
    if year == 2020:
        end_of_seasons.loc[i, 'seas_id'] = 4
    if year == 2019:
        end_of_seasons.loc[i, 'seas_id'] = 3
    if year == 2018:
        end_of_seasons.loc[i, 'seas_id'] = 2
    if year == 2017:
        end_of_seasons.loc[i, 'seas_id'] = 1
end_of_seasons
```

```python
end_of_seasons.to_csv('end_of_seasons_final.csv', index=False)
```
